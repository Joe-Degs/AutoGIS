```python
from autogis import *
```

```python
a = Route('coords_list', origin=[(6.7001, -1.62217), (6.70312, -1.62511)],
          dest=[(6.69791, -1.61829), (6.6971, -1.62252)])
```

```python
a.explore()
```

```python
g = Graph('route_polygon', a)
```

```python
g.download(network_type='drive').nodes_and_edges()
```

```python
g.geometry_from_polygon(key='buildings', tags={'building': True}, facecolor='lightgray', alpha=0.7)
```

```python
g.shortest_path_with_route(a, edge_kwargs=dict(color='red', linewidth=2, linestyle='--'),
                           node_kwargs=dict(markersize=30, color='red'))
```

```python
g.extent()
```

```python
g.static_plot(basemap=True, color='gray', edge_kwargs=dict(linewidth=0.75),
              filepath="images/shortest_path.png", node_kwargs=dict(markersize=2),
              shortest_path=True, geom_keys=['buildings'])
```
