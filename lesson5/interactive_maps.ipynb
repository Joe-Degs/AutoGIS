- ### Adding Layers to Map

```python
cur_loc = folium.Map(location=[6.6721, -1.5694], zoom_start=12,
                     control_scale=True, prefer_canvas=True)

# add marker
folium.Marker(
    location=[6.67514, -1.57303],
    popup='UITS ICTC, KNUST',
    icon=folium.Icon(color='green', icon='ok-sign'),
).add_to(cur_loc)

cur_loc
```

- #### plotting data from geodataframe on folium maps

```python
# File path
points_fp = r"data/addresses.shp"

# Read the data
points = gpd.read_file(points_fp)

#Check input data
points.head()
```

```python
# convert points to geojson
point_gjson = folium.features.GeoJson(points, name="Public transport stations")
```

```python
# Check the GeoJSON features
# point_gjson.data.get('features')
```

```python
# Create a Map instance
m = folium.Map(location=[60.25, 24.8], tiles='cartodbpositron', zoom_start=11, control_scale=True)

# Add points to the map instance
point_gjson.add_to(m)

# Alternative syntax for adding points to the map instance
#m.add_child(points_gjson)

#Show map
m
```

- #### add layer control

```python
folium.LayerControl().add_to(m)

m
```
